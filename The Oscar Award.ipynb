```python
#Importing the basic libraries
import numpy as np
import pandas as pd
import csv
```

```python
awardData = pd.read_csv('the_oscar_award.csv')
awardData.head()
```

```python
imdbData = pd.read_csv('imdbTop250.csv')
imdbData.head()
```

### Creating a new column in awardData:                  

```python
awardData.insert(7, "Number of wins", 0)
```

```python
awardData.head(-1)
```

```python
awardData.info()
```

```python
title = ""
wins = 0

for i in range(0, len(awardData)):
    print(i)
    title = awardData['film'][i]
    if awardData['Number of wins'][i] > 0:
        continue
    if awardData['winner'][i] == True:
        wins += 1
    for j in range(i+1, len(awardData)):
        if awardData['film'][j] == title:
            if awardData['winner'][j] == True:
                wins += 1
    for k in range(0, len(awardData)):
        if awardData['film'][k] == title:
            awardData.at[k,'Number of wins'] = wins
    wins = 0
```
