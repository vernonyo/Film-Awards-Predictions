## Importing the basic libraries:

```python
import numpy as np
import pandas as pd
import csv
```

### Add all the dataframes from the box office website for years 1977-2022 into a list:

```python
list = []

for i in range(1977, 2023):
    url = f'https://www.boxofficemojo.com/year/world/{i}'
    df = pd.read_html(url)
    print(url)
    dfOld = df[0]
    dfOld.insert(3, "Release Year", i)
    list.append(dfOld)
```

### Combining all dataframes into one dataframe:

```python
salesData = pd.concat(list, ignore_index=True)
```
