Essential Libraries

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

Import the Dataset

```python
oscar = pd.read_csv('oscarData.csv')
oscar.head()
```

```python
oscar.info()
```

Exploratory Analysis

1. Exploring the response variable Number of Wins.

```python
# Generations in the Dataset
print("Winners :", len(oscar["Number of Wins"].unique()))

# Pokemons in each Generation
print(oscar["Number of Wins"].value_counts())
sb.catplot(y = "Number of Wins", data = oscar, kind = "count")
```

Check out the Numeric and Categorical predictors separately.

```python
# Numeric Predictors
oscar[['Rating','Sales']].describe()
```

```python
oscar[['Genre']].describe()
```

```python
# Rating
f, axes = plt.subplots(1, 2, figsize=(18, 4))
sb.boxplot(data = oscar['Rating'], orient = "h", ax = axes[0])
sb.histplot(data = oscar['Rating'], ax = axes[1])
```

```python
# Sales
f, axes = plt.subplots(1, 2, figsize=(18, 4))
sb.boxplot(data = oscar['Sales'], orient = "h", ax = axes[0])
sb.histplot(data = oscar['Sales'], ax = axes[1])
```

```python
# Genre of the Movie
sb.catplot(y = 'Genre', data = oscar, kind = "count")
```

Relationship between Rating and Number of Wins

```python
f = plt.figure(figsize=(16, 8))
sb.stripplot(x = 'Rating', y = 'Number of Wins', data = oscar)
```

```python
sb.catplot(x='Rating',y='Number of Wins',data=oscar,kind='box', orient="h",aspect=4)
```

Relationship between Sales and Number of Wins

```python
f = plt.figure(figsize=(16, 8))
sb.stripplot(x = 'Sales', y = 'Number of Wins', data = oscar)
```

```python
sb.catplot(x='Sales',y='Number of Wins',data=oscar,kind='box', orient="h",aspect=4)
```

Relationship between Genre and Number of Wins

```python
f = plt.figure(figsize=(16, 8))
sb.stripplot(x = 'Genre', y = 'Number of Wins', data = oscar)
```

```python
# Rating vs Genre
f = plt.figure(figsize=(6, 4))
sb.heatmap(oscar.groupby(['Number of Wins', 'Genre']).size().unstack(), 
           linewidths = 1, annot = True, fmt = 'g', annot_kws = {"size": 18}, cmap = "BuGn")
```
